```python
import pandas as pd
import numpy as np
```

```python
# Read the data from file
df = pd.read_csv("adult.data.csv") 
df.head()
```

```python
# Find the total number of each race
race_count = df['race'].value_counts()
print(race_count)
```

```python
# Find the average age of men and women
average_age_men = df[df["sex"] == "Male"]["age"]
average_age_men = round(np.mean(average_age_men),1)

average_age_women = df[df["sex"] == "Female"]["age"]
average_age_women = round(np.mean(average_age_women),1)

print(average_age_men, average_age_women, sep="\n")
```

```python
# Find the number of married and non-married
# Find the total number of those in relationship and those that are not 
marital_status = df['marital-status'].value_counts()
relationship = df['relationship'].value_counts()

print(marital_status, relationship, sep="\n")
```

```python
# What is the percentage of people who have a Bachelor's degree?
barchelors = len(df[df["education"] == "Bachelors"])
total_education = len(df["education"])
percentage_barchelors = round(100*(barchelors/total_education),1) 
print(percentage_barchelors)
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
# with and without `Bachelors`, `Masters`, or `Doctorate`

advanced_education = df[df["education"].isin(["Bachelors", "Masters", "Doctorate"])]
normal_education = df[~df["education"].isin(["Bachelors", "Masters", "Doctorate"])]
```

```python
non_percentage_advanced = len(advanced_education[advanced_education.salary == ">50K"])
advanced_education_rich = round(100*(non_percentage_advanced) / len(advanced_education), 1) 
non_percentage_normal = len(normal_education[normal_education.salary == ">50K"])
normal_education_rich = round(100*(non_percentage_normal) / len(normal_education), 1)

print(advanced_education_rich,normal_education_rich,  sep="\n")
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df["hours-per-week"].min()
print(min_work_hours)
```

```python
# What is the maximum number of hours a person works per week (hours-per-week feature)?
max_work_hours = df["hours-per-week"].max()
print(max_work_hours)
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df[df["hours-per-week"] == min_work_hours]
rich_percentage = round(len(num_min_workers[num_min_workers.salary == ">50K"]) / len(num_min_workers)*100, 1)
print(rich_percentage)
```

```python
# What percentage of the people who work the maximum number of hours per week have a salary of >50K?
num_max_workers = df[df["hours-per-week"] == max_work_hours]
poor_percentage = round(len(num_max_workers[num_max_workers.salary == ">50K"]) / len(num_max_workers)*100, 1)
print(poor_percentage)
```

```python
# What country has the highest percentage of people that earn >50K?
new_df  = df.loc[df["salary"] == ">50K"]["native-country"].value_counts()
new_df2 = df["native-country"].value_counts()
richest = (new_df/new_df2).max()
highest_earning_country = (new_df/new_df2).sort_values(ascending = False).index[0] 
highest_earning_country_percentage = round(100*richest, 1)
print(highest_earning_country, highest_earning_country_percentage, sep="\n")
```

```python
# What country has the highest percentage of people that earn <50K
new_df3  = df.loc[df["salary"] == "<50K"]["native-country"].value_counts()
new_df4 = df["native-country"].value_counts()
richest = (new_df3/new_df4).min()
normal_earning_country = (new_df/new_df2).sort_values(ascending = True).index[0] 
normal_earning_country_percentage = round(100*richest, 1)
print(normal_earning_country, normal_earning_country_percentage, sep="\n")
```

```python
# Identify the most popular occupation for those who earn >50K in Germany.
germany = df.loc[(df["salary"] == ">50K") & (df["native-country"] == "Germany")]["occupation"] 
top_IN_occupation = pd.Series.mode(germany)[0]
print(top_IN_occupation)
```
